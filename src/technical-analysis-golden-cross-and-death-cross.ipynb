```python
import datetime as dt
import matplotlib.pyplot as plt
import yfinance as yf

def plot(symbol, period):
    df = yf.Ticker(symbol).history(period=period, interval='1d')
    
    df.Close.plot(alpha=0.25, label='Price')
    df.Close.rolling(window=200).mean().plot(label='200d SMA')
    df.Close.rolling(window=50).mean().plot(label='50d SMA')
    
    plt.title(f"{symbol} ({period})")
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    
    plt.show()
    
plot('DOGE-USD','3y')
```
